# Model equations: cortex, teacher, and cerebellum

This short notebook is meant as a readable entry point.  It does not train a policy.  It shows the action decomposition used in the project: a cortex policy proposes a movement, a teacher policy supplies a corrective target during perturbation windows, and a cerebellum-like module learns the residual between them.

The main supervised target is

$$\Delta u_t^* = u_t^{teacher} - u_t^{ctx},$$

and the executed residual policy is

$$u_t = \mathrm{clip}(u_t^{ctx} + \alpha_t\Delta u_t, u_{min}, u_{max}).$$

The gate $\alpha_t$ is optional.  In the strongest result the important part was not the gate, but the teacher-provided residual target.

```
import numpy as np
rng = np.random.default_rng(7)
action_dim = 6
u_ctx = rng.normal(0, 0.25, size=action_dim)
u_teacher = u_ctx + rng.normal(0, 0.10, size=action_dim)
delta_star = u_teacher - u_ctx
alpha = 0.8
u_final = np.clip(u_ctx + alpha * delta_star, -1.0, 1.0)
print('cortex action:', np.round(u_ctx, 3))
print('teacher action:', np.round(u_teacher, 3))
print('residual target:', np.round(delta_star, 3))
print('executed action:', np.round(u_final, 3))
```

In biological language, this is closer to an error-correction pathway than to a replacement motor cortex.  The residual is only asked to learn the difference between a nominal motor command and a stronger perturbation-trained command.
