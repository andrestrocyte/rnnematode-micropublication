# Video index

The videos are not needed to compute the statistics, but they are useful for seeing what the benchmark is doing.  The index below points to representative nominal, moderate, and severe push episodes for several methods and bodies.

```
from pathlib import Path
import pandas as pd
video_index = Path('videos/20260422_000426/curated/representative_video_index.csv')
df = pd.read_csv(video_index)
print('videos indexed:', len(df))
df[['env','method','tier','kind','severity','curated_path']].head(12)
```

```
for env in sorted(df.env.unique()):
    severe = df[(df.env == env) & (df.tier == 'severe')]
    print('\n' + env)
    for _, row in severe.head(4).iterrows():
        print(f"  {row.method}: {row.curated_path}")
```
