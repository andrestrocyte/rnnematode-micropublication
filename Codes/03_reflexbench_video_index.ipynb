# Video index

The videos are not needed to compute the statistics, but they are useful for seeing what the benchmark is doing.  The packaged CSV is a validation artifact that lists representative nominal, moderate, and severe episodes for several methods and bodies.  The actual MP4 files are optional; if they are included, their paths are listed in the index.

```python
from pathlib import Path
import pandas as pd

def find_submission_root(start=None):
    start = Path.cwd() if start is None else Path(start).resolve()
    for p in [start, *start.parents]:
        if (p / 'video_index').exists() and (p / 'myst_submission').exists():
            return p
        if (p / 'myst.yml').exists() and (p.parent / 'video_index').exists():
            return p.parent
        if (p / 'index.md').exists() and (p.parent / 'video_index').exists():
            return p.parent
    return start

ROOT = find_submission_root()
video_index = ROOT / 'video_index' / 'representative_video_index.csv'
if not video_index.exists():
    raise FileNotFoundError(
        f'Missing video validation index: {video_index}. ' +
        'Place representative_video_index.csv in video_index/ before running this notebook.'
    )
df = pd.read_csv(video_index)
print('video rows indexed:', len(df))
print('MP4 files are optional; this CSV records the validation index used for representative videos.')
df[['env','method','tier','kind','severity','curated_path']].head(12)
```

```python
for env in sorted(df.env.unique()):
    severe = df[(df.env == env) & (df.tier == 'severe')]
    print('\n' + env)
    for _, row in severe.head(4).iterrows():
        print(f"  {row.method}: {row.curated_path}")
```
